```python
def checkKey(t):
    if int(t)>0 and int(t)<26:
        return int(t)
    elif int(t)>26:
        t=int(t)%26
        return int(t)
    else:
        print("Key should be greater than 0")
        return checkKey(input("Enter again:"))

def encryption():
    pt=input("Enter plain text in lowercase:\t")
    plt=pt.upper()
    c=0
    ct=""
    for i in plt:
        if( (ord(i)>=65 and ord(i)<=90) or (ord(i)==32)) :
          c=c+1
          if(c==len(plt)): 
              k=int(input("Enter the key:\t"))
              t=checkKey(k)
              sum=0
              for i in plt.split():
                for j in i:
                  sum=sum+((((ord(j)-65)+t)%26)+65)
                  t=ord(j)-65
                  ct=ct+chr(sum)
                  sum=0
                ct=ct+" "
              print("cipher text is: ",ct)
            
        else:
          print("Invalid character entered!\n")
          encryption()
          break

    return ""

def decryption():
    ct=input("Enter CIPHER TEXT in UPPERCASE:\t")
    clt=ct.upper()
    c=0
    pt=""
    for i in clt:
        if(((ord(i)>=65 and ord(i)<=90)) or (ord(i)==32)):
          c=c+1
          if(c==len(clt)): 
              k=int(input("Enter the key:\t"))
              t=checkKey(k)
              
              sum=0
              for i in clt.split():
                for j in i:
                  sum=sum+((((ord(j)-65)-t)%26)+65)
                  t=ord(j)-65-t
                  pt=pt+chr(sum)
                  sum=0
                pt=pt+" "
              print("cipher text is: ",pt.lower())
            
        else:
          print("Invalid character entered!\n")
          decryption()
          break

    return ""


def toenter():
  print("--------auto key cipher-------")
  print("enter 1 for encryption")
  print("enter 2 for decryption")

  choice=int(input("enter your choice:"))
  if(choice==1):
      print(encryption())
   
  elif(choice==2):
         
      print(decryption())
      
  else:
      print("invalid input,enter again:")
      return toenter()
  return ""

d=toenter()
print(d)

print("\nDo you want to continue?")
print("Enter 'Y' for yes and Enter 'N' for no")
c=input("Enter your choice:\t")
print()

while(c=='Y' or c=='y'):
  toenter()
  print("\nDo you want to continue?")
  print("Enter 'Y' for yes and Enter 'N' for no")
  c=input("Enter your choice:\t")
  
else:
  if(c=='N' or c=='n'):
    print("thank you")
    print("please visit again")
  else:
    print("Invalid character entered!\n")
    toenter()

  
```
