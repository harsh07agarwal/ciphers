```python
import math

def encrypt(s):
    text=input("enter the plain text:")
    result=''
    for i in range(len(text)): 
        char = text[i] 
  
        # Encrypt uppercase characters 
        if (ord(char)>=65 and ord(char)<=90): 
            result += chr((ord(char) * int(s) - 65) % 26 + 65) 
        # Encrypt lowercase characters 
        elif(ord(char)>=97 and ord(char)<=122):
            result += chr((ord(char) * int(s)- 97) % 26 + 97)
        elif(ord(char)==32):
            result+=chr(ord(char))
       
        else:
            print("invalid input,enter again:")
            
            return encrypt(int(s))
            break            
    return result

def decrypt(b):
    t=input("enter the cipher text:")          
    r=""
    print("inverse:",b)    
    for i in range(len(t)): 
        ch = t[i] 
  
        # Decrypt uppercase characters 
        if (ord(ch)>=65 and ord(ch)<=90): 
            r += chr((ord(ch) * int(b)-65) % 26 + 65) 
        # Decrypt lowercase characters 
        elif(ord(ch)>=97 and ord(ch)<=122): 
            r += chr((ord(ch) * int(b) - 97) % 26 + 97)
        elif(ord(ch)==32):
            r+=chr(ord(ch))
        else:
            print("invalid input")
            print("enter again")
            return decrypt(int(b))
            break
    return r     

def checkkey(a):
  if(int(a)<0):
    print("invalid input,key should be greater than 0")
    return checkkey(input("enter again:"))
  else:
    if(math.gcd(int(a),26)==1):
      a=int(a)%26
      return int(a)
    else:
      print("invalid input,since gcd(",a,",26 )!=1")
      return checkkey(input("enter again:"))

def modInverse(a) : 	 
	for x in range(1, 26) : 
		if ((int(a) * x) % 26 == 1) : 
			return x   
	return 1

def toenter():
  print("---------multiplicative cipher-------")
  print("enter 1 for encryption")
  print("enter 2 for decryption")

  choice=int(input("enter your choice:"))
  if(choice==1):
      a = checkkey(int(input("Enter key:")))      
      print("cipher:", encrypt(a))
   
  elif(choice==2):
      a = checkkey(int(input("Enter key:")))  
      b=modInverse(a)    
      print("plain text:", decrypt(b))
      
  else:
      print("invalid input,enter again:")
      return toenter()
  return ""

d=toenter()
print(d)

print("\nDo you want to continue?")
print("Enter 'Y' for yes and Enter 'N' for no")
c=input("Enter your choice:\t")
print()

while(c=='Y' or c=='y'):
  toenter()
  print("\nDo you want to continue?")
  print("Enter 'Y' for yes and Enter 'N' for no")
  c=input("Enter your choice:\t")
  
else:
  if(c=='N' or c=='n'):
    print("thank you")
    print("please visit again")
  else:
    print("Invalid character entered!\n")
    toenter()
    

  
```
